```python
import matplotlib.pyplot as plt 
```

```python
img = cv2.imread("photooo.jpg")
img = cv2.resize (img,None, fx=0.5 , fy= 0.5)
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#plt.imshow(cv2.cvtColor(gray,cv2.COLOR_BGR2RGB))



adaptive_Threshold = cv2.adaptiveThreshold(gray , 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,85,30)
plt.imshow(cv2.cvtColor(adaptive_Threshold,cv2.COLOR_BGR2RGB))
config = "--psm 6"
text = pytesseract.image_to_string(adaptive_Threshold, lang = "eng" ,config = config)
print(text)
```
